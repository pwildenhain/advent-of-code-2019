# Day 8



## Part 1


```python
with open("day-8-input.txt") as f:
    puzzle_input = f.read()
    num_list = [int(num) for num in puzzle_input]
```

```python
width = 25
height = 6
layer_size = width * height
num_layers = int(len(num_list) / layer_size)
```

```python
layers = [[[None for _ in range(width)]  for _ in range(height)] for _ in range(num_layers)]
num_list_idx = 0
for layer in range(num_layers):
    for sub_layer in range(height):
        for idx in range(width):
            layers[layer][sub_layer][idx] = num_list[num_list_idx]
            num_list_idx += 1
```

```python
def flatten(layer):
    return [num for sub_list in layer for num in sub_list]
```

```python
zero_list = [flatten(layers[layer]).count(0) for layer in range(num_layers)]
min_zero_layer = zero_list.index(min(zero_list))
```

```python
flatten(layers[min_zero_layer]).count(1) * flatten(layers[min_zero_layer]).count(2) 
```

## Part 2

```python
final_image = [[2 for _ in range(width)]  for _ in range(height)]
for layer in range(num_layers):
    for sub_layer in range(height):
        for idx in range(width):
            value = layers[layer][sub_layer][idx]
            # Only set value if it's not transparent AND the current spot hasn't already
            # been filled by another layer
            if value != 2 and final_image[sub_layer][idx] == 2:
                final_image[sub_layer][idx] = value
```

```python
print(''.join([str(num) for num in flatten(final_image)]))
```
